```python
import pandas as pd
```

```python
# import and display csv file
original_data = pd.read_csv('taylor_swift_spotify.csv')

original_data
```

```python
# remove unnecessary column
original_data = original_data.drop(columns='Unnamed: 0')

original_data
```

```python
# create list of albums to include in updated dataframe
album_list = ["Taylor Swift", "Fearless (Taylor's Version)", "Speak Now (Deluxe Edition)", "Red (Taylor's Version)",
            "1989 (Deluxe Edition)", "reputation", "Lover", "folklore (deluxe version)", "evermore (deluxe version)",
             "Midnights (3am Edition)"]

# filters out albums not found in album list and displays updated dataframe
updated_data = original_data[original_data['album'].isin(album_list)]

updated_data
```

```python
# drop duplicated song names
updated_data = updated_data.drop_duplicates(subset='name')

# sort by album release date
updated_data = updated_data.sort_values(by=['release_date'])
```

```python
# for proof-reading purposes
updated_data.to_csv('updated_data.csv')
```
