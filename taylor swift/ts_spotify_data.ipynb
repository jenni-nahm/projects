```python
import pandas as pd
```

```python
original_data = pd.read_csv('taylor_swift_spotify.csv')

original_data
```

```python
original_data.drop(columns='Unnamed: 0')
```

```python
album_list = ["Taylor Swift", "Fearless (Taylor's Version)", "Speak Now (Deluxe Edition)", "Red (Taylor's Version)",
            "1989 (Deluxe Edition)", "reputation", "Lover", "folklore (deluxe version)", "evermore (deluxe version)",
             "Midnights (3am Edition)"]

updated_data = original_data[original_data['album'].isin(album_list)]

updated_data
```

```python
updated_data = updated_data.drop_duplicates(subset='name')
```

```python
updated_data.to_csv('updated_data.csv')
```
