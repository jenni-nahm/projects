```python
import pandas as pd
```

```python
books_df = pd.read_csv("Books Read in 2022.csv")

books_df['Words'] = pd.to_numeric(books_df['Words'].replace(',','', regex=True))

books_df
```

```python
books_df.plot.line(y ='Pages', title = 'Book Length in Pages', xlabel = 'Book Number', 
                    ylabel = 'Number of Pages', color = 'r')
```

```python
books_df.plot.line(y ='Words', title = 'Book Length in Words', xlabel = 'Book Number', 
                    ylabel = 'Number of Words', color = 'c')
```

```python
books_df.plot.hist(y = 'Pages', title = 'Book Length in Pages', color = 'r')
```

```python
books_df.plot.hist(y = 'Words', title = 'Book Length in Words', color = 'c')
```

```python
genre_count = books_df.groupby(['Genre'])['Books'].count()
```

```python
genre_count.plot.bar(ylabel = 'Number of Books', title = 'Number of Books by Genre', color = 'g')
```

```python
series_count = books_df.groupby('Series')['Books'].count()
```

```python
series_count.plot.bar(ylabel = 'Number of Books', title = 'Number of Books by Series', color = 'purple')
```

```python
series_none = series_count.index == 'NONE'
series_wo_none = series_count[~series_none]

series_wo_none.plot.bar(ylabel = 'Number of Books', title = 'Number of Books by Series', color = 'purple')
```
